```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
merged = pd.read_csv("../data/processed/merged_1.csv", parse_dates=["date", "intervention_start", "intervention_end"])
```

```python
# Sleep features
```

```python
# --- Sleep Deficit ---
merged["avg_sleep_user"] = merged.groupby("user_id")["sleep_hours"].transform("mean")
merged["sleep_deficit"] = merged["avg_sleep_user"] - merged["sleep_hours"]

# --- Rolling Sleep Average (7-day) ---
merged["sleep_rolling_avg_7d"] = merged.groupby("user_id")["sleep_hours"].transform(lambda x: x.rolling(7, min_periods=3).mean())

# --- Sleep Variability (7-day rolling std) ---
merged["sleep_var_7d"] = merged.groupby("user_id")["sleep_hours"].transform(lambda x: x.rolling(7, min_periods=3).std())

# --- Sleep Quality Normalized ---
merged["sleep_quality_norm"] = merged["sleep_quality"] / merged["sleep_quality"].max()
```

```python
# Large positive sleep_deficit suggests chronic under-sleeping.
# High sleep_var_7d suggests irregular sleep pattern → fatigue risk.
```

```python
# Work and Productivity
```

```python
# --- Workload Index ---
merged["workload_index"] = (
    merged["work_hours"] +
    0.5 * merged["meetings_count"] +
    0.2 * merged["emails_received"]
)

# --- Work Stress Ratio ---
merged["work_stress_ratio"] = merged["workload_index"] / (merged["stress_level"] + 1)

# --- Productivity Score ---
merged["productivity_score"] = merged["tasks_completed"] / (merged["work_hours"] + 1)

# --- Focus-to-Fatigue Ratio ---
merged["focus_fatigue_ratio"] = merged["focus_score"] / (merged["stress_level"] + 1)
```

```python
# low focus_fatigue_ratio even with high work_hours suggests early burnout signal.
```

```python
# Lifestyle and screen time
```

```python
# --- Screen Stress Ratio ---
merged["screen_stress_ratio"] = merged["screen_time_hours"] / (merged["stress_level"] + 1)

# --- Outdoor Time Ratio ---
merged["outdoor_ratio"] = merged["outdoor_time_minutes"] / (merged["outdoor_time_minutes"] + merged["screen_time_hours"] + 1)

# --- Diet Quality Scaled ---
merged["diet_score_norm"] = merged["diet_quality"] / merged["diet_quality"].max()

# --- Calorie Balance ---
merged["calorie_balance"] = merged["calories_intake"] - merged.groupby("user_id")["calories_intake"].transform("mean")
```

```python
# high outdoor ratio = positive recovery
```

```python
# Mood and stress
```

```python
# --- Rolling Averages ---
merged["mood_rolling_avg_7d"] = merged.groupby("user_id")["mood_score"].transform(lambda x: x.rolling(7, min_periods=3).mean())
merged["stress_rolling_avg_7d"] = merged.groupby("user_id")["stress_level"].transform(lambda x: x.rolling(7, min_periods=3).mean())

# --- Mood Volatility (std) ---
merged["mood_volatility_7d"] = merged.groupby("user_id")["mood_score"].transform(lambda x: x.rolling(7, min_periods=3).std())

# --- Stress Trend (difference between last and first 7-day window) ---
merged["stress_trend"] = merged.groupby("user_id")["stress_level"].transform(lambda x: x.diff().rolling(7, min_periods=3).mean())
```

```python
# Health Index (core output variable for viz and LLM insights)
```

```python
# Normalize key features to 0–1 range
for col in ["sleep_deficit", "stress_level", "workload_index", "screen_stress_ratio"]:
    merged[f"{col}_norm"] = (merged[col] - merged[col].min()) / (merged[col].max() - merged[col].min())

# Compute the index (lower stress/deficit, higher mood/quality)
merged["health_twin_index"] = (
    0.3 * (1 - merged["sleep_deficit_norm"]) +
    0.25 * (1 - merged["stress_level_norm"]) +
    0.25 * merged["mood_score"] / merged["mood_score"].max() +
    0.2 * merged["sleep_quality_norm"]
) * 100
```

```python
# Intervention Impact (Allows causal analysis)
```

```python
merged["under_intervention"] = merged["intervention_type"].notna().astype(int)

# Rolling improvement in mood during interventions
merged["mood_change_during_intervention"] = merged.groupby("user_id")["mood_score"].transform(lambda x: x.diff()) * merged["under_intervention"]

# Post-intervention recovery metric
merged["stress_reduction_post_intervention"] = merged.groupby("user_id")["stress_level"].transform(lambda x: x.shift(1) - x) * merged["under_intervention"]
```
